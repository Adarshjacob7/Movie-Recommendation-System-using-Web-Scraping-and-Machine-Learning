```python
import pandas as pd
import glob
import os

# Folder that contains your CSV files
folder_path = r"E:\Miscellaneous\Data Science\Python\Projects\2 Web scrapping Boxoffice\merge files"

# Columns you want to keep
required_columns = [
    "movie_title", "imdb_rating", "director", "cast",
    "imdb_votes", "certification", "genres", "description"
]

# List to store dataframes
dfs = []

# Read all CSV files in the folder
for file in glob.glob(os.path.join(folder_path, "*.csv")):
    print(f"Reading: {file}")
    df = pd.read_csv(file)

    # Keep only the required columns that exist in file
    available_cols = [col for col in required_columns if col in df.columns]
    df = df[available_cols]

    dfs.append(df)

# Merge all dataframes
merged_df = pd.concat(dfs, ignore_index=True)

# Save the merged result
output_path = os.path.join(folder_path, "merged_movies.csv")
merged_df.to_csv(output_path, index=False)

print("Done! Merged file saved as:", output_path)
```
